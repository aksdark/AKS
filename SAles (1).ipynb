```python
x=df['Branch'].value_counts()
plt.pie(x,labels=['A','B','C'],startangle=90,explode=[0.1,0.1,0.1])
plt.title('Distribution of Branch')
plt.legend(x)
```

```python
x=df['Branch']
sb.countplot(x)
plt.xlabel('branch')
plt.title('Representing no.of branches')
```

```python
x=df['Payment'].value_counts()
```

```python
x=df['Payment'].value_counts()
plt.pie(x,labels=['Cash','Ewallet','Credit card'],startangle=90,explode=[0.1,0.2,0.3])
plt.title('Distribution of payment methods')
plt.legend(x)
```

```python
y=df['Product line']
x=df['Quantity']
sb.barplot(x,y)
plt.xlabel('Quantity')
plt.ylabel('Product line')
plt.title('Distribution of product line')
```

```python
sb.barplot(data = df, x = "Product line", y = "gross income", hue = "Gender")
plt.xticks(rotation = 45)
plt.title('distribution of gross income of product')
plt.show()
```

```python
df.hist(figsize = (15, 7))
```

```python
sb.countplot(data= df, x='City', hue= 'Branch')
plt.title('no.of branches in different cities')
```

```python
x=df['City'].value_counts()
plt.pie(x,labels=['Yangon','Naypyitaw','Mandalay'],startangle=90,explode=[0.1,0.1,0.1],radius=0.8)
plt.title('Distribution of payment methods')
plt.legend(x)
```

```python
y=df['Tax 5%']
sb.histplot(y)
```

```python
sb.scatterplot(data=df,x='Quantity',y='gross income',hue='Product line')
plt.rcParams["figure.figsize"]=(5,10)
plt.show()
```
